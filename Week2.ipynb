```python
import cv2 as cv
import matplotlib.pyplot as plt

# Membaca gambar dalam warna asli
img = cv.imread('/content/drive/MyDrive/PCVK/kucing.jpg')

# Membuat citra baru hanya dengan saluran Red-Blue
red_blue = img.copy()
red_blue[:, :, 1] = 0  # Set saluran hijau menjadi 0 (hanya tinggalkan saluran merah dan biru)

# Menampilkan citra dalam saluran Red-Blue
plt.imshow(cv.cvtColor(red_blue, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Red-Blue Channel')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from skimage import io

# Membaca gambar

image = io.imread('/content/drive/MyDrive/PCVK/orang.jpg')

# Koordinat wajah (contoh)
x_face, y_face, width_face, height_face = 250, 150, 200, 200

# Membuat plot
fig, ax = plt.subplots(1)

# Menampilkan gambar asli
ax.imshow(image)

# Membuat rectangle pada wajah
face_rectangle = patches.Rectangle((x_face, y_face), width_face, height_face, linewidth=2, edgecolor='r', facecolor='none')
ax.add_patch(face_rectangle)

# Koordinat pusat dan radius circle (contoh)
x_center, y_center, radius = 350, 240, 100
# Membuat circle pada wajah
face_circle = patches.Circle((x_center, y_center), radius, linewidth=2, edgecolor='g', facecolor='none')
ax.add_patch(face_circle)

# Menampilkan gambar dengan rectangle dan circle
plt.show()
```
